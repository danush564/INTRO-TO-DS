#Removing NULL Values

subset -> will remove the complete row of the mentioned column
how = all -> is to remove null values if entire column is null
how = any -> is to remove null values if any 1 column value is null

```python
cleaned_any = df.dropna(how='all')
cleaned_any
```

```python
cleaned_all = df.dropna(how='any')
print(cleaned_all)
```

#Finding Duplicates

```python
import numpy as np
data = {
    'Name' : ['John','Sarah','John','Emily','David'],
    'Age' : [25,31,25,35,27],
    'Gender' : ['M','F','M','F','M'],
    'Salary' : [50000,70000,50000,80000,55000]
    }
df = py.DataFrame(data)
print(df.duplicated())
print("\nNot Duplicated\n")
print(~df.duplicated())
```

#Filtering

```python
data = {
    'Name' : ['John','Sarah','Mike','Emily','David'],
    'Age' : [25,31,29,35,27],
    'Gender' : ['M','F','M','F','M'],
    'height' : [1.6,1.78,1.65,1.83,1.7]
    }
df = py.DataFrame(data)
filtered = df[(df['Gender'] == 'F') | (df['Gender'] == 'M') & (df['height']>=1.7)]
filtered
```

#a.  select the rows where clients with primary education have subscribed to a deposit?


---



```python
df.loc[(df['education'] == 'primary') & (df['deposit'] == 'yes')]
```

#b.  select the rows where clients who have not subscribed to a deposit?


---



```python
df.loc[df['deposit'] == 'no']
```

#c. select the rows where clients who have subscribed to a deposit either have a housing or a personal loan?


---



```python
df.loc[(df['deposit'] == 'yes') & ((df['housing'] == 'yes') | (df['loan'] == 'yes'))]
```

#d. select the rows where clients with secondary education who have not subscribed to a deposit?


---



```python
df.loc[(df['education']== 'secondary') & (df['deposit'] == 'no')]
```

#e. select the rows where  clients who have subscribed to a term deposit as an outcome of the successful marketing campaign?


---



```python
df.loc[(df['poutcome'] == 'success') & (df['deposit'] == 'yes')]
```

#f. select the rows where unemployed clients who have not subscribed to deposit?


---



```python
df.loc[(df['job'] == 'unemployed') & (df['deposit'] == 'yes')]
```
